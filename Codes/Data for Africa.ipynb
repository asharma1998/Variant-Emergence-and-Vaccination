```python
import pandas as pd

# Read the dataset from a CSV file
#file_path = "path/to/your/dataset.csv"  # Replace with the actual path to your dataset file
df = pd.read_csv(r'C:\Users\sharm\Documents\My Folder\Project_SARS\locations.csv')

```

```python
# Create categories
rna_vaccines = ["Pfizer/BioNTech", "Moderna"]
adenovirus_vaccines = ["Oxford/AstraZeneca", "Janssen", "Sputnik V", "Sputnik Light", "Convidecia", "iNCOVACC"]
inactivated_virus_vaccines = ["Sinopharm BIBP", "CoronaVac", "Covaxin", "Valneva", "Sinopharm WIBP"]
subunit_vaccines = ["Novavax", "Sanofi-GSK", "Abdala", "EpiVacCorona", "Zifivax", "Soberana 02"]

# Function to categorize vaccines
def categorize_vaccine(vaccine):
    if any(brand.lower() in vaccine.lower() for brand in rna_vaccines):
        return "RNA Vaccines"
    elif any(brand.lower() in vaccine.lower() for brand in adenovirus_vaccines):
        return "Adenovirus Vector Vaccines"
    elif any(brand.lower() in vaccine.lower() for brand in inactivated_virus_vaccines):
        return "Inactivated Virus Vaccines"
    elif any(brand.lower() in vaccine.lower() for brand in subunit_vaccines):
        return "Subunit Vaccines"
    else:
        return "Uncategorized"

# Apply categorization to the 'vaccines' column
df['vaccine_category'] = df['vaccines'].apply(categorize_vaccine)

# Display the categorized DataFrame
print(df[['location', 'vaccines', 'vaccine_category']])
```
